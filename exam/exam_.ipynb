```python
import cv2
import matplotlib.pyplot as plt

# Upload the image
image = cv2.imread("DJI_0073.JPG", cv2.IMREAD_GRAYSCALE)

# Show the image
plt.imshow(image, cmap="gray")
plt.title("Drone Image")
plt.axis("off")
plt.show()
```

```python
# matriz of the image
print(image)
print("Shape:", image.shape)  # (rows, cols, channels)
```

```python
#upload of the template
template = cv2.imread("template.JPG", cv2.IMREAD_GRAYSCALE)
```

```python
# Show the template
plt.imshow(template, cmap="gray")
plt.title("Template Image")
plt.axis("off")
plt.show()
```

Compute the correlation using Opencv, the method used was TM_SQDIFF which uses the same formula as the paper Fast Template Matching

```python
import time
import cv2 as cv

start_time = time.time()
# Compute the cross-correlation
correlation = cv.matchTemplate(image,template,cv.TM_SQDIFF)
end_time = time.time()

print(f"Execution time: {end_time - start_time} seconds")
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(correlation, cmap='gray')
plt.title("Correlation result")
plt.axis('off')
plt.show()
```

Compute the normalized with the method TM_CCORR_NORMED

```python
start_time = time.time()
# Compute the cross-correlation normalized 
normalized_correlation = cv.matchTemplate(image,template,cv.TM_CCORR_NORMED)
end_time = time.time()

print(f"Execution time: {end_time - start_time} seconds")
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(normalized_correlation, cmap='gray')
plt.title("Correlation result")
plt.axis('off')
plt.show()
```

```python
# Compute the integral image
integral_image = cv.integral(image)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(integral_image, cmap='gray')
plt.title("Integral Image Result")
plt.axis('off')
plt.show()
```
