```python
import os 
import numpy as np
import matplotlib.pyplot as plt
import cv2
import time
```

```python
img = cv2.imread('/Users/angelicamariamorenorojas/Desktop/Master/Image processing/DJI_0073.JPG')
img
```

```python
# fourier transform the img 
img_fft = np.fft.fft2(img)
img_fft_shift = np.fft.fftshift(img_fft)
img_fft_shift_log = np.log(1 + np.abs(img_fft_shift))
```

```python
img_fft_shift_log.min(), img_fft_shift_log.max()
```

```python
img_fft_shift_log
```

```python
# plot the fft
plt.figure(figsize=(10, 10))
plt.imshow(img_fft_shift_log[:,:,0], cmap='gray')
plt.title('Fourier Transform of the Image')
```
